# Numpy

NumPy (Numerical Python) is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays.

```python
import numpy as np
```

```python
# Create a NumPy array
arr = np.array([1, 2, 3, 4, 5])
print("NumPy array:", arr)
```

```python
# Get the type of the array
print("Type of array:", type(arr))

# Get the shape of the array
print("Shape of array:", arr.shape)
```

```python
# Create a 2D array (matrix)
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("\n2D array:\n", matrix)

# Get the shape of the 2D array
print("Shape of 2D array:", matrix.shape)
```

You can perform various operations on NumPy arrays, such as element-wise operations, mathematical functions, and linear algebra operations.

```python
# Element-wise operations
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print("Addition:", arr1 + arr2)
print("Multiplication:", arr1 * arr2)

# Mathematical functions
print("Square root:", np.sqrt(arr))
print("Sine:", np.sin(arr))
```

NumPy also provides functions to create arrays with specific values, like zeros, ones, or a range of numbers.

```python
# Create an array of zeros
zeros_arr = np.zeros((2, 3))
print("\nArray of zeros:\n", zeros_arr)
```

```python
# Create an array of ones
ones_arr = np.ones((3, 2))
print("\nArray of ones:\n", ones_arr)
```

```python
# Create an array with a range of numbers
range_arr = np.arange(10)
print("\nArray with range:", range_arr)

# Create an array with a range of numbers with a step
range_step_arr = np.arange(0, 10, 2)
print("Array with range and step:", range_step_arr)
```

```python
# Matrix multiplication
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

print("\nMatrix multiplication:\n", matrix1 @ matrix2) # Using the @ operator
print("Matrix multiplication (dot product):\n", np.dot(matrix1, matrix2)) # Using np.dot

# Transpose of a matrix
print("\nTranspose of matrix:\n", matrix1.T)

# Inverse of a matrix (if square and non-singular)
try:
    print("\nInverse of matrix:\n", np.linalg.inv(matrix1))
except np.linalg.LinAlgError:
    print("\nMatrix is singular, cannot compute inverse.")
```

NumPy's `np.linalg.solve` function can be used to solve systems of linear equations of the form Ax = B, where A is a square matrix, x is the vector of unknowns, and B is the constant vector.

```python
# Solve the system of linear equations:
# 2x + 3y = 12
# x + y = 5

# Define the coefficient matrix A
A = np.array([[2, 3], [1, 1]])

# Define the constant vector B
B = np.array([12, 5])

# Solve for x (the vector of unknowns)
x = np.linalg.solve(A, B)

print("Solution for the system of linear equations (x, y):", x)
```

```python
import numpy as np

arr = np.array([0, np.pi/2, np.pi])

# Trigonometric functions
print("Sine:", np.sin(arr))
print("Cosine:", np.cos(arr))
print("Tangent:", np.tan(arr))

# Logarithmic functions
arr2 = np.array([1, np.e, np.e**2])
print("\nNatural logarithm:", np.log(arr2)) # natural log (base e)
print("Base 10 logarithm:", np.log10(arr2)) # base 10 log
```

```python
# Time series of stock prices
stock_prices = np.array([150.25, 152.75, 153.50, 155.00, 154.25])

# Compute the mean stock price
mean_price = np.mean(stock_prices)

# Compute the cumulative sum of stock prices
cumulative_sum = np.cumsum(stock_prices)

print("Mean Stock Price:", mean_price)
print("Cumulative Sum of Stock Prices:", cumulative_sum)
```

```python
# Example of Fast Fourier Transform (FFT)
# Create a simple signal (sum of two sine waves)
import numpy as np
import matplotlib.pyplot as plt

# Parameters
sampling_rate = 1000  # Hz
duration = 1          # seconds
frequency1 = 5        # Hz
frequency2 = 15       # Hz

# Time vector
t = np.linspace(0, duration, sampling_rate * duration, endpoint=False)

# Create the signal
signal = 0.5 * np.sin(2 * np.pi * frequency1 * t) + 0.2 * np.sin(2 * np.pi * frequency2 * t)

# Perform FFT
fft_result = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(signal), 1/sampling_rate)

# Take the absolute value of the FFT result and consider only positive frequencies
fft_magnitude = np.abs(fft_result)[:len(signal)//2]
positive_frequencies = frequencies[:len(signal)//2]

# Plot the original signal
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(t, signal)
plt.title('Original Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# Plot the FFT result
plt.subplot(1, 2, 2)
plt.plot(positive_frequencies, fft_magnitude)
plt.title('FFT Result')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a sample time series dataset for stock prices
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
stock_prices = np.random.rand(100) * 100 + 150  # Generating some random prices around 150

stock_df = pd.DataFrame({'Date': dates, 'Price': stock_prices})

# Set 'Date' column as the index
stock_df.set_index('Date', inplace=True)

# Plot the stock prices
plt.figure(figsize=(12, 6))
stock_df['Price'].plot()
plt.title('Sample Stock Price Time Series')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```
